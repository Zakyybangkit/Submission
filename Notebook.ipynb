## Visualization & Explanatory Analysis

### Pertanyaan 1:

````python
# Menghitung jumlah cnt dengan kategori holiday=0
jumlah_cnt_holiday_0 = Day_df.loc[Day_df['holiday'] == 0, 'cnt'].sum()

# Menghitung jumlah cnt dengan kategori holiday=1
jumlah_cnt_holiday_1 = Day_df.loc[Day_df['holiday'] == 1, 'cnt'].sum()

print(f"Jumlah cnt dengan kategori holiday=0: {jumlah_cnt_holiday_0}")
print(f"Jumlah cnt dengan kategori holiday=1: {jumlah_cnt_holiday_1}")

# Data untuk bar chart
kategori = ['Holiday=0', 'Holiday=1']
jumlah_cnt = [jumlah_cnt_holiday_0, jumlah_cnt_holiday_1]

# Membuat bar chart
plt.bar(kategori, jumlah_cnt, color=['skyblue', 'salmon'])
plt.xlabel('Kategori Holiday')
plt.ylabel('Jumlah cnt')
plt.title('Jumlah cnt berdasarkan Kategori Holiday')
plt.show()
````

### Pertanyaan 2:

````python
# Memilih variabel independen (features)
X = Day_df[['temp', 'hum', 'windspeed']]

# Variabel dependen (target)
y = Day_df['cnt']

# Menambahkan konstanta untuk model regresi
X = sm.add_constant(X)

# Membuat model regresi linear
model = sm.OLS(y, X).fit()

# Menampilkan hasil regresi
print(model.summary())

# Scatter plot untuk variabel temp
plt.scatter(Day_df['temp'], Day_df['cnt'], label='Temperature', alpha=0.7)

# Scatter plot untuk variabel hum
plt.scatter(Day_df['hum'], Day_df['cnt'], label='Humidity', alpha=0.7)

# Scatter plot untuk variabel windspeed
plt.scatter(Day_df['windspeed'], Day_df['cnt'], label='Windspeed', alpha=0.7)

plt.xlabel('Variabel Independen')
plt.ylabel('Variabel Dependen (cnt)')
plt.title('Hubungan Variabel Independen dengan cnt')
plt.legend()
plt.show()
````
