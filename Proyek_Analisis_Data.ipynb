## Visualization & Explanatory Analysis

### Pertanyaan 1:

````python
# Menghitung jumlah cnt berdasarkan kategori weathersit (1-4)
cnt_by_weathersit = Day_df.groupby('weathersit')['cnt'].sum()

# Membuat bar chart
plt.bar(cnt_by_weathersit.index, cnt_by_weathersit.values, color='skyblue')
plt.xlabel('Weathersit')
plt.ylabel('Jumlah cnt')
plt.title('Jumlah cnt berdasarkan Kategori Weathersit')
plt.show()
````

### Pertanyaan 2:

````python

# Menghitung korelasi antara dua variabel (misalnya, temp dan cnt)
correlation_temp_cnt = Day_df['temp'].corr(Day_df['cnt'])
correlation_hum_cnt = Day_df['hum'].corr(Day_df['cnt'])
correlation_windspeed_cnt = Day_df['windspeed'].corr(Day_df['cnt'])
print(f"Korelasi antara variabel temp dan cnt: {correlation_temp_cnt:.2f}")
print(f"Korelasi antara variabel hum dan cnt: {correlation_hum_cnt:.2f}")
print(f"Korelasi antara variabel windspeed dan cnt: {correlation_windspeed_cnt:.2f}")

# Menghitung korelasi antara variabel temp, hum, windspeed dengan cnt
correlation_matrix = Day_df[['temp', 'hum', 'windspeed', 'cnt']].corr()

# Visualisasi scatter plot untuk temp terhadap cnt
plt.figure(figsize=(8, 6))
sns.scatterplot(x='temp', y='cnt', data=Day_df, alpha=0.7)
plt.xlabel('Temperature (Â°C)')
plt.ylabel('cnt')
plt.title('Scatter Plot: Temperature vs cnt')
plt.show()

# Visualisasi scatter plot untuk hum terhadap cnt
plt.figure(figsize=(8, 6))
sns.scatterplot(x='hum', y='cnt', data=Day_df, alpha=0.7)
plt.xlabel('Humidity')
plt.ylabel('cnt')
plt.title('Scatter Plot: Humidity vs cnt')
plt.show()

# Visualisasi scatter plot untuk windspeed terhadap cnt
plt.figure(figsize=(8, 6))
sns.scatterplot(x='windspeed', y='cnt', data=Day_df, alpha=0.7)
plt.xlabel('Windspeed')
plt.ylabel('cnt')
plt.title('Scatter Plot: Windspeed vs cnt')
plt.show()
````
